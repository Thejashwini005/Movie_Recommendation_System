```python
print(movies.columns.tolist())
```

```python
movies=movies[['Title', 'Year of Release', 'Rating', 'Number of Reviews',
       'Popularity Index', 'Movie Cast', 'Director', 'Description']]
```

```python
movies
```

```python
movies = movies.copy()

```

```python
movies
```

```python
new_data = movies
```

```python
new_data
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features=10000, stop_words='english')
```

```python
cv
```

```python
vector=cv.fit_transform(new_data['Movie Cast'].values.astype('U')).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vector)
```

```python
similarity
```

```python
new_data[new_data['Title']=="Vikram"].index[0]
```

```python
distance = sorted(list(enumerate(similarity[2])), reverse=True, key=lambda vector:vector[1])
for i in distance[0:5]:
    print(new_data.iloc[i[0]].Title)
```

```python
def recommand(movies):
    index=new_data[new_data['Title']==movies].index[0]
    distance = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda vector:vector[1])
    for i in distance[0:5]:
        print(new_data.iloc[i[0]].Title)
```

```python
recommand("Vikram")
```
